## O(1) Algo 

With 0(1) we will always have a function that will return an  object after one unit of working time, and 
the time will always be the same irrespective often we change our list.

```python
def get_first(data):
    x = data[0]**data[0]+data[0]**data[0]
    return x
```

```python
def array(data):
    start = time.time()
    get_first(data)
    end = time.time()
    return end-start
```

## O(N) Algo 

```python
def raise_power(data):
    v = []
    for i in data:
        v.append(i**2)
    return v
```

```python
def array(data,func):
    start = time.time()
    raise_power(data)
    end = time.time()
    return end-start
```

```python
On_vectors_len = []
On_vectors_time = []
for i in range(10000,1000000,10000):
    v = np.arange(i)
    time_ = array(v,raise_power)
    On_vectors_len.append(len(v))
    On_vectors_time.append(time_)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9, 6)
plt.plot(On_vectors_len[3:],On_vectors_time[3:])
plt.ylabel('Execution Time (sec)')
plt.xlabel('Length of list')
plt.title('Time complexity')
#plt.ylim(0, 0.005)
plt.legend(['O(n)'])
```

## $O(n^2)$  Algo  

```python
def even(data,constants=[1,2]):
    v = []
    w = []
    for i in data:
        i= i**2
        for k in constants:
            if k%2==0:
                w.append([i**k])
            else:
                w.append(i**k)
    return v
```

```python
def array(data,func):
    start = time.time()
    func(data)
    end = time.time()
    return end-start
```

```python
ot_vectors_len = []
ot_vectors_time = []
for i in range(10000,1000000,10000):
    v = np.arange(i)
    time_ = array(v,even)
    ot_vectors_len.append(len(v))
    ot_vectors_time.append(time_)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9, 6)
plt.plot(ot_vectors_len,ot_vectors_time)
plt.ylabel('Execution Time (sec)')
plt.xlabel('Length of list')
plt.title('Time complexity')
#plt.ylim(0, 0.000005)
plt.legend(['O(n^2)'])
```

## O(log(n))

No need to go through all elements

```python
def frac_count(x): 
    if len(x)<30:
        x=x[0]
        return x
    
    
```

```python
lg_vectors_len = []
lg_vectors_time = []
for i in range(10000,1000000,10000):
    v = np.arange(i)
    time_ = array(v,frac_count)
    lg_vectors_len.append(i)
    lg_vectors_time.append(time_)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9, 6)
plt.plot(lg_vectors_len[:],lg_vectors_time[:])
plt.ylabel('Execution Time (sec)')
plt.xlabel('Length of list')
plt.title('Time complexity')
plt.ylim(0, 0.00005)
plt.legend(['O(log(n)'])
```

## Recursive  Functions

```python
def fact(n):
    if n==2:
        return 1
    else :
        return n*fact(n-1)
fact(5)
```
